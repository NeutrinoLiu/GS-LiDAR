```python

import shutil
import os
import opencood.utils.pcd_utils as pcd_utils

src_dir = os.path.join("advCP", "data", "OPV2V")
dest_dir = os.path.join("data", "opv2v")

mode = "test"

os.makedirs(os.path.join(dest_dir, mode), exist_ok=True)

for scenario in os.listdir(os.path.join(src_dir, mode)):
    os.makedirs(os.path.join(dest_dir, mode, scenario), exist_ok=True)
    for vehicle in os.listdir(os.path.join(src_dir, mode, scenario)):
        if vehicle.endswith(".yaml"):
            shutil.copy(os.path.join(src_dir, mode, scenario, vehicle), os.path.join(dest_dir, mode, scenario, vehicle))
            print(f"Copied scene {scenario}, vehicle {vehicle}")
        else:
            os.makedirs(os.path.join(dest_dir, mode, scenario, vehicle), exist_ok=True)
            for frame in os.listdir(os.path.join(src_dir, mode, scenario, vehicle)):
                if frame.endswith(".pcd"):
                    arr = pcd_utils.pcd_to_np(os.path.join(src_dir, mode, scenario, vehicle, frame))
                    arr.tofile(os.path.join(dest_dir, mode, scenario, vehicle, frame.replace(".pcd", ".bin")))
                    print(f"Converted scene {scenario}, vehicle {vehicle}, frame {frame} to {frame.replace('.pcd', '.bin')}")
                else:
                    shutil.copy(os.path.join(src_dir, mode, scenario, vehicle, frame), os.path.join(dest_dir, mode, scenario, vehicle, frame))
                    print(f"Copied scene {scenario}, vehicle {vehicle}, frame {frame}")
            print(f"Processed scene {scenario}, vehicle {vehicle}")
    print(f"Processed scene {scenario}")
```
